### Задание №3
Реализовать методы сортировки строк числовой матрицы в соответствии с
заданием. Оценить время работы каждого алгоритма сортировки и сравнить его со
временем стандартной функции сортировки. Испытания проводить на сгенерированных

```python
# Сортировка выбором.
import copy
import random
import time

start_time = time.time()

def generate_matrix(rows, cols, min_val=0, max_val=100):

    matrix = []
    for _ in range(rows):
        row = [random.randint(min_val, max_val) for _ in range(cols)]
        matrix.append(row)
    return matrix

def selection_sort_matrix(matrix):

  rows = len(matrix)
  cols = len(matrix[0])
  N = rows * cols


  flat_list = [matrix[i][j] for i in range(rows) for j in range(cols)]


  for i in range(N - 1):
    min_val = flat_list[i]
    min_index = i
    for j in range(i + 1, N):
      if flat_list[j] < min_val:
        min_val = flat_list[j]
        min_index = j

    if min_index != i:
      flat_list[i], flat_list[min_index] = flat_list[min_index], flat_list[i]


  sorted_matrix = []
  for i in range(0, N, cols):
    sorted_matrix.append(flat_list[i:i + cols])

  return sorted_matrix


rows = 4
cols = 4
min_value = 2
max_value = 19

my_matrix = generate_matrix(rows, cols, min_value, max_value)

print("Исходная матрица:")
for row in my_matrix:
    print(row)

sorted_matrix = selection_sort_matrix(my_matrix)

print("\nОтсортированная матрица (построчно):")
for row in sorted_matrix:
    print(row)
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка обменом. || Сортировка пузырьком.
start_time = time.time()

import random
import copy


def generate_matrix(rows, cols, min_val=0, max_val=100):
    matrix = []
    for _ in range(rows):
        row = [random.randint(min_val, max_val) for _ in range(cols)]
        matrix.append(row)
    return matrix


def bub_sor(A):
    for j in range(len(A) - 1):
        for i in range(len(A) - 1 - j):
            if A[i] > A[i + 1]:
                A[i], A[i + 1] = A[i + 1], A[i]


rows = int(input("Введите количество строк матрицы: "))
cols = int(input("Введите количество столбцов матрицы: "))
user_n = int(input("Введите размер подматрицы для вывода (N): "))


mtx = generate_matrix(rows, cols)
print("Исходная матрица:")
for row in mtx:
    print(row)


cop4 = copy.deepcopy(mtx)  
cop5 = []
for i in cop4:
    for i2 in i:
        cop5.append(i2)
N = len(cop5)
bub_sor(cop5)


sorte = []
for i in range(0, N, user_n):
    sorte.append(cop5[i:i + user_n])


print("\nСортировка пузырьком:")
for row in sorte:
    print(row)

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка Шелла.
start_time = time.time()
import random

def generate_matrix(rows, cols, min_val=0, max_val=100):
    matrix = []
    for _ in range(rows):
        row = [random.randint(min_val, max_val) for _ in range(cols)]
        matrix.append(row)
    return matrix


def shell_sort(arr):

    n = len(arr)
    gap = n // 2

    while gap > 0:
        for i in range(gap, n):
            temp = arr[i]
            j = i
            while j >= gap and arr[j - gap] > temp:
                arr[j] = arr[j - gap]
                j -= gap
            arr[j] = temp
        gap //= 2
    return arr


def print_matrix(matrix):
    for row in matrix:
        print(row)

rows = 3
cols = 3
min_value = 0
max_value = 100

    
random_matrix = generate_matrix(rows, cols, min_value, max_value)

print("Исходная матрица:")
print_matrix(random_matrix)


    
sorted_matrix = []
for row in random_matrix:
    sorted_row = shell_sort(row)
    sorted_matrix.append(sorted_row)

print("\nМатрица с отсортированными строками (сортировка Шелла):")
print_matrix(sorted_matrix)


print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Быстрая сортировка.
start_time = time.time()
import random

def generate_matrix(rows, cols, min_val=0, max_val=100):
    matrix = []
    for _ in range(rows):
        row = [random.randint(min_val, max_val) for _ in range(cols)]
        matrix.append(row)
    return matrix

def print_matrix(matrix):
    for row in matrix:
        print(row)

def quicksort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]  
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quicksort(left) + middle + quicksort(right)

rows = 5
cols = 4
min_value = 0
max_value = 100

random_matrix = generate_matrix(rows, cols, min_value, max_value)
print("Исходная матрица:")
print_matrix(random_matrix)

flat_list = [element for row in random_matrix for element in row]

sorted_list = quicksort(flat_list)

sorted_matrix = []
for i in range(rows):
    sorted_matrix.append(sorted_list[i*cols:(i+1)*cols])

print("\nОтсортированная матрица:")
print_matrix(sorted_matrix)

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
    
```

```python
# Турнирная сортировка
import heapq
import time
import random

start_time = time.time()


def tournament_sort(data):
    size = len(data)
    if size <= 1:
        return data 

    heap = []
    for i in range(0, size, 2):
        if i + 1 < size:
            heap.append(min(data[i], data[i + 1]))
        else:
            heap.append(data[i]) 

    heapq.heapify(heap)

    sorted_list = []
    indices = {}

    for element in heap:
        indices[element] = [i for i, x in enumerate(data) if x == element][
            0] 

    while heap:
        winner = heapq.heappop(heap)
        sorted_list.append(winner)

        winner_index = indices[winner]
        pair_index = winner_index ^ 1 

        if pair_index < size:
            try:
                index_in_heap = heap.index(winner)  
            except ValueError:
                index_in_heap = -1

            if index_in_heap >= 0:
                loser = data[pair_index]

                heapq.heappush(heap, loser)
                indices[loser] = pair_index

    return sorted_list


def generate_matrix(rows, cols, min_val, max_val):
    matrix = []
    for _ in range(rows):
        row = [random.randint(min_val, max_val) for _ in range(cols)]
        matrix.append(row)
    return matrix

rows = 5
cols = 5
min_val = 1
max_val = 100

random_matrix = generate_matrix(rows, cols, min_val, max_val)

print("Original Matrix:")
for row in random_matrix:
    print(row)

flattened_list = [element for row in random_matrix for element in row]

sorted_data = tournament_sort(flattened_list)

print("\nSorted Flattened List:")
print(sorted_data)

print("--- {0} ms ---".format(round((time.time() - start_time) * 1000)))
```

```python
# Пирамидальная сортировка
import time
import random
import numpy as np

start_time = time.time()

def heapify(arr, n, i):
  largest = i  
  left = 2 * i + 1 
  right = 2 * i + 2 

  if left < n and arr[i] < arr[left]:
    largest = left

  if right < n and arr[largest] < arr[right]:
    largest = right

  if largest != i:
    arr[i], arr[largest] = arr[largest], arr[i]

    heapify(arr, n, largest)


def heap_sort(arr, matrix_size=None):
  n = len(arr)
  if matrix_size is not None:
      matrix_rows, matrix_cols = matrix_size
      print("Исходная матрица:")
      print(np.array(arr).reshape((matrix_rows, matrix_cols)))
      print("-" * 20)


  for i in range(n // 2 - 1, -1, -1): 
    heapify(arr, n, i)
    if matrix_size is not None:
        print(f"После heapify {i} - формирование Max-кучи:")
        print(np.array(arr).reshape((matrix_rows, matrix_cols)))
        print("-" * 20)

  for i in range(n - 1, 0, -1):
    arr[0], arr[i] = arr[i], arr[0]
    if matrix_size is not None:
        print(f"Меняем arr[0] и arr[{i}]")
        print(np.array(arr).reshape((matrix_rows, matrix_cols)))
        print("-" * 20)

    heapify(arr, i, 0)
    if matrix_size is not None:
        try:
            print(f"После heapify (размер кучи {i}):")
            print(np.array(arr[:i]).reshape((matrix_rows, matrix_cols)))
            print("-" * 20)

        except ValueError:
            print(f"  Невозможно отобразить кучу размера {i} в матрицу {matrix_rows}x{matrix_cols}")
            print(f"  Текущий размер массива arr[:{i}]: {len(arr[:i])}")
            print("-" * 20)
  return arr

start_time = time.time()
matrix_rows = 3
matrix_cols = 4
matrix_size = (matrix_rows, matrix_cols)
arr = [random.randint(1, 99) for _ in range(matrix_rows * matrix_cols)]

print("Размер матрицы:", matrix_size)
sorted_arr = heap_sort(arr, matrix_size)
print("Отсортированная матрица (как одномерный массив):", sorted_arr)

print("--- {0} ms ---".format(round((time.time() - start_time) * 1000)))
```
